```python
import pandas as pd

df = pd.read_excel (r'Data\Tweets#portugal.xlsx', sheet_name='Folha1', header= None)
df1 = pd.read_excel (r'Data\Tweets#portugal1.xlsx', sheet_name='Folha1', header= None)
df2 = pd.read_excel (r'Data\Tweets#portugal2.xlsx', sheet_name='Folha1', header= None)
```

```python
dfportugal = pd.concat([df, df1,df2], ignore_index=True)
```

```python
dfportugal.head()
```

```python
dfportugal.columns = ['user', 'tweet', 'link', 'date']
```

```python
dfportugal.head()
```

```python
#we'll need worldcloud to create it and matplotlib to plot it
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

#our text to feed the wordcloud comes from the snippet collumn
text = dfportugal.tweet.values

#let's create the wordcloud
wordcloud = WordCloud(
    width = 4000,
    height = 3000,
    background_color = 'black',
    stopwords = STOPWORDS).generate(str(text))

#with the wordcloudcreated it's time to plot it using matplotlib
fig = plt.figure(
    figsize = (40, 30),
    facecolor = 'k',
    edgecolor = 'k')
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis('off')
plt.tight_layout(pad=0)

#let's see? 
plt.show()
```
